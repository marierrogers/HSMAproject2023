# Explore independent and dependent variables

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url_annual = 'annual_modelling_data.csv'
url_monthly = 'monthly_modelling_data.csv'
df_annual = pd.read_csv(url_annual,parse_dates=['month_year'],dayfirst=True)
df_monthly = pd.read_csv(url_monthly,parse_dates=['month_year'],dayfirst=True)
```

```python
corr_cols = ['leaver_rate', 'sickness_absence',
                                   'local_unemployment','joiner_rate','anxiety_stress']
correlation_matrix = df_annual[corr_cols].corr()
# Create correlation matrix heatmap using Seaborn
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```

```python
sorted(df_annual['staff_group'].unique())
```

```python
staff_group = 'Nurses & health visitors'
df_sg = df_annual[df_annual['staff_group'] == staff_group].copy()
```

```python
correlation_matrix = df_sg[corr_cols].corr()
# Create correlation matrix heatmap using Seaborn
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```

```python
df_sg
sns.set()

plt.figure(figsize=(10, 6))  
sns.lineplot(data=df_sg, x='month_year', y='leaver_rate', hue='staff_group')
plt.xlabel('Date')
plt.ylabel('Rate')
plt.title('Rate over Time by Staff Group')
plt.legend(title='Staff Group', loc='best')  
plt.xticks(rotation=45)  
plt.show()
```
